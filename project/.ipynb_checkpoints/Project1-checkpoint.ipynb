```python
import pandas as pd
df = pd.read_csv('CLIP-35L33G-50.pileup', sep='\t', header=None)
```

```python
df
```

```python
import re
from scipy.stats import entropy
from collections import Counter

# retain the longest isoforms among overlapping transcripts
def base_filter(base):
    return re.sub('[^(A|T|G|C)]', '', base.upper())

# Crosslinking-induced reverse-transcription error score (CRES) was determined by using Shannon’s entropy
bases = ['A', 'T', 'G', 'C']
def get_CRES(seq):
    sequence_length = len(seq)
    counter = Counter(seq)
    probs = [counter[base] / sequence_length for base in bases]
    return entropy(probs, base=2)
```

```python
df['filtered_seq'] = df[4].map(base_filter)
df_filtered = df[df['filtered_seq']!=''].reset_index(drop=True)
df_filtered['CRES'] = df_filtered['filtered_seq'].map(get_CRES)
```

```python
df_filtered
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(df_filtered[df_filtered['CRES']>0]['CRES'])
plt.show()
```

```python
df_filtered[df_filtered['CRES']>0]['CRES'].describe()
```

```python
df_filtered[df_filtered['CRES']>0][3].describe()
```

```python
# we used constant CRES cut-offs (e.g., 0.8) together with depth cut-offs (e.g.,50 reads) for other analyses
df_positive = df_filtered[(df_filtered['CRES']>1) & (df_filtered[3]<300)]
df_positive
```

```python
# generate negative dataset
import random

base_dict = {0:'A', 1:'T', 2:'G', 3:'C'}
def generate_random_seq(n):
    return ''.join([base_dict[random.randint(0,3)] for _ in range(n)])

seq_len = list(df_positive[3])
sns.kdeplot(seq_len)
plt.show()

positive_seqs = list(df_positive['filtered_seq'])
negative_seqs = [generate_random_seq(l) for l in seq_len]
```

```python
# interagte positive and negative datasets
df_full = pd.DataFrame({'seq':positive_seqs+negative_seqs, 'Y':[1]*len(positive_seqs)+[0]*len(negative_seqs)})
df_full
```

```python
df_full.to_csv('./dataset.csv', index=False)
```
